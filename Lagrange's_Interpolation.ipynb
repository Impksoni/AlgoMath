**A Python program for Lagrange's Interpolation**

```python
# Importing the Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the Lagrange Interpolation Function

def lagrange_interpolation(x_values, y_values, x_to_interpolate):
    """
    Perform Lagrange Interpolation for given data points.

    Parameters:
    x_values : list or array of x coordinates
    y_values : list or array of y coordinates
    x_to_interpolate : the x value at which to evaluate the interpolation

    Returns:
    float : interpolated y value
    """
    n = len(x_values)
    result = 0.0

    for i in range(n):
        term = y_values[i]
        for j in range(n):
            if i != j:
                term *= (x_to_interpolate - x_values[j]) / (x_values[i] - x_values[j])
        result += term

    return result
```

```python
# Example 1 - Basic Interpolation

# Known data points
x_vals = [1, 2, 4]
y_vals = [2, 3, 1]

# Interpolate at x = 3
x_interp = 3
y_interp = lagrange_interpolation(x_vals, y_vals, x_interp)

print(f"Interpolated value at x = {x_interp} is y = {y_interp}")
```

```python
# Example 2 - Interpolation at Multiple Points

# Interpolate at several points between 1 and 4
x_test = np.linspace(1, 4, 100)
y_test = [lagrange_interpolation(x_vals, y_vals, x) for x in x_test]

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(x_test, y_test, label="Lagrange Interpolation Curve", color='blue')
plt.scatter(x_vals, y_vals, color='red', label="Known Data Points")
plt.axvline(x=3, linestyle='--', color='gray', label="Interpolated x = 3")
plt.title("Lagrange Interpolation Example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 3 - Using more points (e.g., function approximation)

# Approximate sin(x) using interpolation at 5 points
x_vals = [0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi]
y_vals = [np.sin(x) for x in x_vals]

x_interp = np.pi / 3
y_interp = lagrange_interpolation(x_vals, y_vals, x_interp)

print(f"Example 3: Interpolated sin({x_interp:.3f}) ≈ {y_interp:.5f}")
print(f"Actual sin({x_interp:.3f}) = {np.sin(x_interp):.5f}")
```

```python
# Plot sin(x) interpolation

x_test = np.linspace(0, np.pi, 200)
y_test_interp = [lagrange_interpolation(x_vals, y_vals, x) for x in x_test]
y_actual = np.sin(x_test)

plt.figure(figsize=(8, 5))
plt.plot(x_test, y_test_interp, label="Interpolated Curve", linestyle='--', color='black')
plt.plot(x_test, y_actual, label="Actual sin(x)", color='lightgreen')
plt.scatter(x_vals, y_vals, color='red', label="Interpolation Points")
plt.title("Interpolating sin(x) using Lagrange")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
